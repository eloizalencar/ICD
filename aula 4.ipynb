Utilização de Grafico com uma biblioteca em python: Usando o matplotib 
https://matplotlib.org/stable/gallery/index (site usado)


```python
# bibliotecas de graficos
import matplotlib.pyplot as plt

#Biblioteca de estruturas de dados (dataframe)
import pandas as pd
import seaborn as sns
```

```python
#dados
pesos = [100, 87, 65, 52]
alturas = [1.89, 1.73, 1.7, 1.64]
```

```python
# grafico de dispersão altura x peso 
plt.scatter(
    alturas, pesos,
    s=120,
    color="#621708"
    )

#codigo para colocar titulo
plt.title(
    "Relação altura x peso/nde pacientes HC-UFTM",
    fontweight="bold",
    fontsize=16
)

#rotulo eixo x
plt.xlabel(
    "Altura",
    fontsize=14
)


#rotulo eixo y
plt.ylabel(
    "Peso",
    fontsize=14
)

#colocar gride
plt.grid

#mostrar grafico
plt.show
```

```python
df=pd.DataFrame({
    "nome": ["Bia", "Ted", "Bob", "Leo"],
    "peso": [100, 87, 65, 52],
    "altura": [1.83, 1.73, 1.7, 1.64],
    "Cor/Raça": ["Branca", "Branca", "Preta","Preta"]})
df.head(1)
```

```python
sns.scatterplot(data=df, x="altura", y="peso", 
                hue="Cor/Raça", s=1203,palette=["#621708", "#bc3908"]   
)

#Titulo
plt.title(
    "Relação altura x peso/nde pacientes HC-UFTM",
    fontweight="bold",
    fontsize=16
)

#rotulo eixo x
plt.xlabel(
    "Altura",
    fontsize=14
)


#rotulo eixo y
plt.ylabel(
    "Peso",
    fontsize=14
)

#colocar gride
plt.grid("on")

#lista de ticks y da gride
plt.yticks(list(range(50, 101, 5)))

#mostrar grafico
plt.show()
```
